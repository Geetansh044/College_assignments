#### Question-1     Write a python program that accepts a list from user and print the alternate element of list. For Example:
How many elements you want to enter? 6 <br>
Enter 6 elements <br>
10
20
30
40
50
60 <br>
Alternate elements are: <br>
10
30
50

```python
n = int(input("How many elements you want yo enter?"))
print("Enter ", n, ' elements')
elements = []
for i in range(n):
    element = int(input())
    elements.append(element)
print("Alterante elements are:")
for i in range(0, len(elements), 2):
    print(elements[i])
```

#### Question-2 Write a python program that accepts a list from user. Your program should reverse the content of list and display it. Do not use reverse() method. 

```python
elements = list(map(int, input("Enter the list elements ").split(' ')))
reversed_elements = [0]*len(elements)

for i in range(len(elements)):
    reversed_elements[len(elements)-i-1] = elements[i]
    
print(reversed_elements)
```

#### Question-3 Find and display the largest number of a list without using built-in function max(). Your program should ask the user to input values in list from keyboard 

```python
elements = list(map(int, input("Enter the list elements ").split(' ')))
max_element = elements[0]

for i in elements:
    if i>max_element:
        max_element = i
print(max_element)
```

#### Question-4 Write a python program that rotates the element of a list so that the element at the first index moves to the second index, the element in the second index moves to the third index, etc., and the element in the last index moves to the first index.

```python
elements = [1,3,5,6,8,20,12,15,12]
last_value = elements[len(elements)-1]

for i in range(len(elements)-1,0, -1):
    elements[i] = elements[i-1]

elements[0] = last_value

print(elements)
```

#### Question-5 Write a program that keeps name and birthday in a dictionary as key-value pairs. The program should display a menu that lets the user search a person’s birthday, add a new name and birthday, change an existing birthday, and delete an existing name and birthday

```python
birthdays = {'Abhishek':'2-April', 'Rahul':'1-Sep', 'Himanshu':'14-July'}

exit = False

while not exit:
    print("""
    Menu:
    1) Search a person's bithday
    2) Add a new name and birthday
    3) Change an existing birthday
    4) Remove a name and birthday
    5) Display all
    6) Exit""")
    
    value = int(input())
    
    if value == 1:
        name = input("Enter the name \n")
        answer = birthdays[name]
        print(answer)
    
    if value == 2:
        name = input("Enter the name \n")
        day = input("Enter the birthday \n")
        birthdays[name] = day
    
    if value == 3:
        name = input("Whose birthday to change \n")
        day = input("Enter the new birthday \n")
        birthdays[name] = day
        
    if value == 4:
        name = input("Whose birthday to remove \n")
        birthdays.pop(name)
        
    if value == 5:
        print(birthdays)
    
    if value == 6:
        break
```

#### Question-6 Write a program that keeps student's name and his marks in a dictionary as key-value pairs. The program should store records of 10 students and display :
        a. The name of all the students
        b. The marks of all the students
        c. Students name and marks of five students in decreasing order of marks obtained.

```python
names_and_marks = {'Abhishek':100, "Rahul":88, "Piyush":93, "Himanshu":76, "Anshul":83, "Anmol":68,
                  "Arjun":99, "Tushar":96, "Anuroop":74, "Pranav":97}

print("Name of all the students \n", list(names_and_marks.keys()), '\n')
print("Marks of all the students \n", list(names_and_marks.values()), '\n')

values = []

def takeSecond(elem):
    return elem[1]

for i,j in names_and_marks.items():
    to_add = [str(i),int(j)]
    values.append(to_add)
    
values.sort(key=takeSecond, reverse=True)

print(values[0:5])
```
